```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
```

```python
np.random.seed(0)
x = np.arange(-1, 1, 0.1)
y = -x**2 + np.random.normal(0, 0.05, len(x))
```

```python
x_train = x[:12]
y_train = y[:12]
```

```python
x_test = x[12:]
y_test = y[12:]
```

```python
def polynomial_fit(degree, x_train, y_train):
    coeffs = np.polyfit(x_train, y_train, degree)
    return np.poly1d(coeffs)
```

```python
def plot_polyfit(degree, x_train, y_train, x_test, y_test):
    poly = polynomial_fit(degree, x_train, y_train)
    x_dense = np.linspace(min(x_train), max(x_train), 200)
    y_dense = poly(x_dense)
    plt.figure()
    plt.scatter(x_train, y_train, label="Entraînement", color="blue")
    plt.scatter(x_test, y_test, label="Test", color="red")
    plt.plot(x_dense, y_dense, label=f"Polynôme degré {degree}")
    plt.legend()
    plt.title(f"Ajustement polynômial degré {degree}")
    plt.show()
```

```python
plot_polyfit(1, x_train, y_train, x_test, y_test)
plot_polyfit(7, x_train, y_train, x_test, y_test)
plot_polyfit(11, x_train, y_train, x_test, y_test)
```
