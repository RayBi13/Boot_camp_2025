4. Am√©liorer le Mod√®le

Suggestions d‚ÄôAm√©lioration

    Ajouter des couches cach√©es et augmenter le nombre de neurones :

```python
model_improved = tf.keras.Sequential([
    tf.keras.layers.Dense(8, input_shape=(2,), activation='relu'),
    tf.keras.layers.Dense(4, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')
])
```

```python
model_improved.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

Entra√Æner pour plus d‚Äô√©poques :

```python
history_improved = model_improved.fit(X, y, epochs=100, verbose=1)
```

5. Visualiser la Fronti√®re de D√©cision

```python
# Impl√©mentation d‚Äôune Fonction plot_decision_boundary

# Pour visualiser comment le mod√®le s√©pare les classes, vous pouvez d√©finir une fonction qui :

#     Cr√©e une grille de points couvrant l‚Äôespace des caract√©ristiques.
#     Pr√©dit la classe pour chaque point de la grille.
#     Affiche un contour de la fronti√®re de d√©cision.

def plot_decision_boundary(model, X, y, title="Fronti√®re de d√©cision"):
    # D√©finir les limites de la grille
    x_min, x_max = X[:, 0].min() - 0.1, X[:, 0].max() + 0.1
    y_min, y_max = X[:, 1].min() - 0.1, X[:, 1].max() + 0.1
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 200),
                         np.linspace(y_min, y_max, 200))

    # Pr√©dire pour chaque point de la grille
    grid = np.c_[xx.ravel(), yy.ravel()]
    preds = model.predict(grid)
    preds = preds.reshape(xx.shape)

    # Tracer la fronti√®re et les points
    plt.figure(figsize=(6,6))
    plt.contourf(xx, yy, preds, levels=[0, 0.5, 1], alpha=0.2, colors=['blue', 'red'])
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis', edgecolor='k')
    plt.title(title)
    plt.xlabel("Feature 1")
    plt.ylabel("Feature 2")
    plt.show()
```
