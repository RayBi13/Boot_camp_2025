```python
ventes_par_etat = df.groupby('State')['Sales'].sum().reset_index()
ventes_par_etat.head()
```

```python
# Visualisation SchÃ©matique de mon atten
plt.figure(figsize=(12, 6))
plt.bar(ventes_par_etat['State'], ventes_par_etat['Sales'], color='skyblue')
plt.xlabel('Ã‰tat')
plt.ylabel('Ventes')
plt.title('Ventes par Ã‰tat')
plt.xticks(rotation=90)  # Rotation des noms des Ã‰tats pour une meilleure lisibilitÃ©
plt.show()
```

```python
# Filtrage dees donnÃ©es pour New York et la Californie
ny_data = df[df['State'] == 'New York']
ca_data = df[df['State'] == 'Californie']

# Calcul des ventes et bÃ©nÃ©fices pour chaque Ã‰tat
ny_ventes_total = ny_data['Sales'].sum()
ca_ventes_total = ca_data['Sales'].sum()

ny_benefices_total = ny_data['Profit'].sum()  
ca_benefices_total = ca_data['Profit'].sum()


```

```python
# Calcule  de la diffÃ©rence
difference_ventes = ny_ventes_total - ca_ventes_total
difference_benefices = ny_benefices_total - ca_benefices_total

# Le rÃ©sultats
print(f"DiffÃ©rence des ventes entre New York et la Californie: {difference_ventes}")
print(f"DiffÃ©rence des bÃ©nÃ©fices entre New York et la Californie: {difference_benefices}")
```
