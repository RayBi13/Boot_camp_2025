```python
# Exo 3
# Visualisation des ventes mensuelles avec un tableau de barres
categories = ['Janvier', 'Février', 'Mars', 'Avril', 'Mai', 'Juin', 'Juillet', 'Août', 'Septembre', 'Octobre', 'Novembre', 'Décembre']
values = [5000, 5500, 6200, 7000, 7500, 8000, 8500, 9000, 9500, 10000, 10500, 11000]
# x = [1, 2, 3, 4]
# y = [10, 20, 25, 30]

plt.bar(categories, values, color='green', edgecolor='black')

plt.title('Ventes mensuelles')
x = plt.xlabel('Month')
y = plt.ylabel('Sales Amount ($)')
plt.show()
```

```python
data = {'CGPA': [8.0, 8.5, 9.0, 9.5, 10.0],}
df = pd.DataFrame(data)

plt.hist(df, bins=30, color='red',edgecolor='black')
plt.title("CGPA")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
# Exo 5
data = {
    'Anxiety': ["True", "False"],
    'Gender': ['M', 'F']
        }
        

df = pd.DataFrame(data)
plt.figure(figsize=(10, 5))
sns.countplot(x='Gender', hue= 'Anxiety', data=df)
plt.title("Stress")
plt.xlabel("Do you have Anxiety?")
plt.ylabel("Choose your gender")
plt.show()
```

```python
# dExo 6
Panic_attacks = {
    'Age' : [20, 30, 40, 50],
    'Do you have Panic Attacks?': ['Yes', 'No', 'Yes', 'No']
}

df = pd.DataFrame(Panic_attacks)
print(df)

# Convertir Yes/No to 1/0 for regression
df['Do you have Panic Attacks?'] = df['Do you have Panic Attacks?'].map({'Yes': 1, 'No': 0})

sns.lmplot(x='Age', y='Do you have Panic Attacks?', data=df, aspect=1.5, height=5)
plt.title('Relationship Between Age and Panic Attacks')
plt.show()
```
