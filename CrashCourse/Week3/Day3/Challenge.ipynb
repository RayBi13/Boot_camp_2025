```python
# Trouver les survivants
df['Survivors'] = df['Aboard'] - df['Fatalities']
```

```python
    # Handling missing values (filling with median/mode or dropping rows/columns as appropriate)

df.dropna(subset=['Fatalities', 'Survivors'], inplace=True)
df.fillna({'Region': 'Unknown'}, inplace=True)

    # Convert date columns to datetime format
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')     

# return df

    
```

```python
# 2. Exploratory Data Analysis
def exploratory_data_analysis(df):
    print("Basic Statistics:")
    print(df[['Fatalities', 'Survivors']].describe())
    
    # Number of crashes over time
    df['Year'] = df['Date'].dt.year
    yearly_crashes = df.groupby('Year').size()
    plt.figure(figsize=(10, 5))
    sns.lineplot(x=yearly_crashes.index, y=yearly_crashes.values)
    plt.title("Number of Crashes Over Time")
    plt.xlabel("Year")
    plt.ylabel("Number of Crashes")
    plt.show()


```

```python
# 3. Statistical Analysis
def statistical_analysis(df):
    print("Statistical Analysis:")
    print("Mean Fatalities:", np.mean(df['Fatalities']))
    print("Median Fatalities:", np.median(df['Fatalities']))
    print("Standard Deviation:", np.std(df['Fatalities']))
    
    # Hypothesis Test: Comparing fatalities between two decades
    df['Decade'] = (df['Year'] // 10) * 10
    fatalities_by_decade = df.groupby('Decade')['Fatalities'].mean()
    
    decade1 = df[df['Decade'] == 1990]['Fatalities'].dropna()
    decade2 = df[df['Decade'] == 2000]['Fatalities'].dropna()
    t_stat, p_value = stats.ttest_ind(decade1, decade2)
    print(f"T-test between 1990s and 2000s fatalities: t-stat = {t_stat}, p-value = {p_value}")

```

```python
# 4. Visualization
def visualize_data(df):
    plt.figure(figsize=(10, 5))
    sns.histplot(df['Fatalities'], bins=30, kde=True)
    plt.title("Distribution of Fatalities")
    plt.xlabel("Fatalities")
    plt.ylabel("Frequency")
    plt.show()
    
    plt.figure(figsize=(12, 6))
    sns.barplot(x='Decade', y='Fatalities', data=df, estimator=np.mean)
    plt.title("Average Fatalities by Decade")
    plt.xticks(rotation=45)
    plt.show()

```
