```python
import numpy as np
import pandas as pd

# données synthétiques de température pour 10 villes sur 12 mois
temperature_data = np.random.uniform(low=-5, high=35, size=(10, 12))


print(temperature_data)
```

```python
# Creatioon DataFrame

# ville names and month names
cities = ['ville1', 'ville2', 'ville3', 'ville4', 'ville5', 'ville6', 'ville7', 'ville8', 'ville9', 'ville10']
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

# Create DataFrame using the synthetic data
df = pd.DataFrame(temperature_data, index=cities, columns=months)

# Display the DataFrame
print(df)
```

```python
# Calculer la température moyenne annuelle pour chaque ville
annual_avg_temp = df.mean(axis=1)


print(annual_avg_temp)
```

```python
# Ville avec la température moyenne la plus élevée
city_highest_avg_temp = annual_avg_temp.idxmax()

# Ville avec la température moyenne la plus basse
city_lowest_avg_temp = annual_avg_temp.idxmin()


print(f"Ville avec la température moyenne la plus élevée: {city_highest_avg_temp}")
print(f"Ville avec la température moyenne la plus élevée: {city_lowest_avg_temp}")
```

```python
import matplotlib.pyplot as plt

# Plotage
plt.figure(figsize=(10, 6))

for city in df.index:
    plt.plot(df.columns, df.loc[city], label=city)

plt.title('Températures moyennes mensuelles pour chaque ville')
plt.xlabel('Mois')
plt.ylabel('Temperature (°C)')
plt.xticks(rotation=45)
plt.legend(title='Ville', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
```

```python
# Conclusions :

# 1. Ville avec la température moyenne la plus élevée :ville 6

# 2. Ville avec la température moyenne la plus basse : Ville 4

# 3. Tendances intéressantes :
# - Certaines villes connaissent des variations saisonnières importantes, avec des températures hautes pendant les mois d'été (juin, juillet et août) et baissant pendant les mois d'hiver (décembre, janvier et février).
# - D'autres villes peuvent afficher des températures plus stables tout au long de l'année.
# - les villes proches de l'équateur ont des températures annuelles plus élevées,
# - Celles situées sous den dessous ont des températures annuelles plus basses.
```
